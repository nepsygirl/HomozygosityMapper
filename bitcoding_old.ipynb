```python
#00	0  00000000
#AA	1  00000001
#CC	2  00000010
#GG	3  00000011
#TT	4  00000100
#ii 5  00000101

#AC	6  00000110
#AG	7  00000111
#AT	8  00001000
#Ai	9  00001001
#CG	10 00001010
#CT	11 00001011
#Ci	12 00001100
#GT	13 00001101
#Gi	14 00001110
#Ti	15 00001111

gt_data = {}  # declare empty dictionary

# allele_data is supposed to hold the columns REF and ALT from the VCF
# allele_data["alt"] is supposed to hold the list of alternative alleles. i.e. "A,G,CCG,T etc"
allele_data = {}  # declare empty dictionary
allele_data["alt"] = {}
allele_data["ref"] = "T" # mockup vcf line with ref "A"


allele_data["alt"][1] = 'GC' # mockup vcf line with first alt allele
allele_data["alt"][2] = 'GCGGGGGTTTTT' # mockup vcf line with second alt allele, man kann noch mehrere kombination von allele hier 
                                        #weiter  hinzufügen
allele_data["alt"][3] = 'ACGT'
allele_data["alt"][4] = 'G'
allele_data["alt"][5] = 'GCG'  #usw

# TODO Mockup VCF use all / for delimiter, but | is also possible
#gt_data["GT"] = "0/1"  # mockup  vcf line 
gt_data["GT"] = "4/5"

#mockup vcf line with indel Ci
#allele_data['ref'] = 'GCTGCGT' #um zu testen ob Ai funktioniert
                               
#allele_data['alt'] = '' #alt is empty when it's a deletion 
#genotyp is 1/1 and alt is empty when its ii 
recoded_genotypes = [] # declare empty list
if gt_data["GT"] == "0/0": # homozygote, both reference allele
    if allele_data["ref"] == "A":
        recoded_genotypes.append('00000001')  
    elif allele_data["ref"] == "C":
        recoded_genotypes.append('00000010')
    elif allele_data["ref"] == "G":
        recoded_genotypes.append('00000011')
    elif allele_data["ref"] == "T":
        recoded_genotypes.append('00000100')
        
elif gt_data["GT"][0] == gt_data["GT"][2]: # homozygote, both identical alternative allele, get indices of alternative alleles and check them if they are equal
    index_of_alt_allele = int(gt_data["GT"][0])
    if allele_data["alt"][index_of_alt_allele] == "A":
        recoded_genotypes.append('00000001')
        
    elif allele_data["alt"][index_of_alt_allele] == "C":
        recoded_genotypes.append('00000010')
        
    elif allele_data["alt"][index_of_alt_allele] == "G":
        recoded_genotypes.append('00000011')
        
    elif  allele_data["alt"][index_of_alt_allele] == "T":
        recoded_genotypes.append('00000100')
        
    elif allele_data["alt"][index_of_alt_allele] == "": ##TODO: indels funktion und wenn es eine deletion in beide  chromosome ist
                                                    #(DELETION)
        recoded_genotypes.append('00000101')

        
 #AB hier Heterozygote Kombinatio, man darf annehmen, dass beide allele dieser Zeile nicht identisch sein können      
elif gt_data["GT"][0] != gt_data["GT"][2]: # heterozygote
    gt_allele_1 = int(gt_data["GT"][0]) # i.e. 2<--/1
    gt_allele_2 = int(gt_data["GT"][2]) # i.e. 2/-->1
    
    if gt_allele_1 == 0:
        allele_1_string = allele_data["ref"]
    else:
        allele_1_string = allele_data["alt"][gt_allele_1] # referencing via gt_allele, which holds index for alternative allele, if it's an alternative allele

    if gt_allele_2 == 0:
        allele_2_string = allele_data["ref"]
        
    else:
        allele_2_string = allele_data["alt"][gt_allele_2]  # referencing via gt_allele, which holds index for alternative allele, if it's an alternative allele
    
    if (allele_1_string == "A" and allele_2_string == "C") or (allele_1_string == "C" and allele_2_string == "A"): #ab 5 waren die kombination heterozygosität und AC und CA sind gleich 
        recoded_genotypes.append('00000110')
  
        
    elif (allele_1_string == "A" and allele_2_string == "G") or (allele_1_string == "G" and allele_2_string == "A"):
        recoded_genotypes.append('00000111')
    
    elif (allele_1_string == "A" and allele_2_string == "T") or (allele_1_string == "T" and allele_2_string == "A"):
        recoded_genotypes.append('00001000')  
        
    elif (allele_1_string == "A" and len(allele_2_string)!=1) or (allele_2_string == "A" and len(allele_1_string)!=1): # If it starts with the letter "A", it's a deletion Ai
        recoded_genotypes.append('00001001') 
        #TODO: indelin tabele auch  abschpeichern 
    
        
    elif (allele_1_string == "C" and allele_2_string == "G") or (allele_1_string == "G" and allele_2_string == "G"):
        recoded_genotypes.append('00001010')
        
    elif (allele_1_string == "C" and allele_2_string == "T") or (allele_1_string == "T" and allele_2_string == "C"):
        recoded_genotypes.append('00001011')
        
#    elif str(allele_data["ref"]).startswith("C") and allele_data["alt"] == "":  # If it starts with the letter "C", it's Ci
    elif (allele_1_string == "C" and len(allele_2_string)!=1) or (allele_2_string == "C" and len(allele_1_string)!=1): # If it starts with the letter "A", it's a deletion Ai
        recoded_genotypes.append('00001100')
        #TODO: indelin tabele auch  abschpeichern
  
    elif (allele_1_string == "G" and allele_2_string == "T") or (allele_1_string == "T" and allele_2_string == "G"):
        recoded_genotypes.append('00001101')
        
#    elif str(allele_data["ref"]).startswith( "G") and allele_data["alt"] == "":  # If it starts with the letter "G", it's Gi
    elif (allele_1_string == "G" and len(allele_2_string)!=1) or (allele_2_string == "G" and len(allele_1_string)!=1): # If it starts with the letter "A", it's a deletion Ai
        recoded_genotypes.append('00001110')
        
        #TODO: indelin tabele auch  abschpeichern
#    elif str(allele_data["ref"]).startswith("T") and allele_data["alt"] == "":  # If it starts with the letter "T", it's Ti
    elif (allele_1_string == "T" and len(allele_2_string)!=1) or (allele_2_string == "T" and len(allele_1_string)!=1): # If it starts with the letter "A", it's a deletion Ai
        recoded_genotypes.append('00001111')
        #TODO: indelin tabele auch  abschpeichern
        
else:
    recoded_genotypes.append('00000000')
print(recoded_genotypes)
    
#funktion um die Indels handeln 
```
